```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Receipts

```python
receipts=pd.read_json("../0.data/receipts.json",lines=True)
receipts['_id'] = receipts['_id'].apply(lambda x: x['$oid'])
```

### Brands

```python
brands=pd.read_json("../0.data/brands.json",lines=True)
brands['_id'] = brands['_id'].apply(lambda x: x['$oid'])
```

### Users

```python
users=pd.read_json("../0.data/users.json",lines=True)
users['_id'] = users['_id'].apply(lambda x: x['$oid'])
```

```python
receipts.head()
```

```python
receipts = receipts.explode('rewardsReceiptItemList')
receipts.reset_index(inplace=True)
```

```python
receipts['_id'] = receipts['_id'].apply(lambda x: x['$oid'])
receipts.head()
```

```python
reciept_items=pd.json_normalize(receipts["rewardsReceiptItemList"])
reciept_items.head()
```

```python
null_percentage = reciept_items.isnull().mean() * 100

null_percentage_filtered = null_percentage[null_percentage > 50]
```

### Around 27 columns out of 34 have more than 50% null data in recipient_items

```python
plt.figure(figsize=(12, 8))
null_percentage_filtered.plot(kind='bar')
plt.title('Percentage of Null Values in Each Column (greater than 50%)')
plt.xlabel('Columns')
plt.ylabel('Percentage of Null Values (%)')
plt.show()
```
